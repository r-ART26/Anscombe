```python
df.groupby('group').describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='x', y='y', hue='group', palette='viridis')
plt.title('Gráfico de dispersión de los 4 grupos')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.grid(True)
plt.legend(title='Grupo')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Obtener los nombres únicos de los grupos
groups = df['group'].unique()

# Iterar sobre cada grupo y crear un gráfico de dispersión con línea de regresión
for group in groups:
    plt.figure(figsize=(6, 5))
    sns.regplot(data=df[df['group'] == group], x='x', y='y', ci=None, scatter_kws={'s': 50}) # Añade la línea de regresión
    plt.title(f'Gráfico de dispersión con línea de regresión para el Grupo {group}')
    plt.xlabel('Eje X')
    plt.ylabel('Eje Y')
    plt.grid(True)
    plt.show()
```
